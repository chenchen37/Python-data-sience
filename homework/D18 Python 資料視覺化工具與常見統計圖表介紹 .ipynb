```python
# 1.畫出 cos 圖檔，並儲存
import random
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)
y = np.cos(x)

plt.plot(x,y)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('cos')

plt.savefig('homework.png',dpi=300,format='png')
```

```python
# 2.給定散點圖顏色
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

C = np.arctan2(Y,X)

plt.scatter(X,Y,s=80,alpha=0.5,c=C)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show
```
